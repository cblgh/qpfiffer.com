```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Import muscles
might = "lifts.csv"
weights = pd.read_csv(might, parse_dates=["Date"], infer_datetime_format=True)
```

```python
weights['Lift'] = weights['Lift'].astype('category')
weights['Total'] = weights['Reps'] * weights['Sets']

# Plot type of lifts
weights['Lift'].value_counts().plot(kind='pie', title='All Squats')
plt.show()
```

```python
# Some summary statistics. Bro, nice lifts.
weights.describe()
```

Gosh, that's a lot of squats. 

```python
weights[weights['Lift'] == 'back-squats'].sum()
```

You've performed a lot of squats since you started tracking them in March 2016. 

```python
# Numbers of days tracked.
td = weights['Date'][1] - weights['Date'][len(weights.index)-1]
-td.days
```

```python
# That's an average of this many squats per day:
weights['Total'][weights['Lift'] == 'back-squats'].sum() / -td.days
```

Let's look a distribution of weights by lift type. Before I even make this plot I'm willing to guess that your squat game is on point.

```python
plt.clf()

# You do better on box squats than back-squats on average. I'm sure this makes sense,
# but I am unaware of the difference and disinclined to look.
weights.groupby('Lift').mean()['Weight (LBS)'].plot(kind='bar')
plt.show()
```

All this is well and good, but numbers should always be considered in terms of distributions - let's look at your squat distribution.

```python
plt.clf()

# Plot back-squat distribution.
weights[weights['Lift'] == 'back-squats']['Weight (LBS)'].plot(kind='hist')
plt.show()
```

That is not a normal distribution. You should be ashamed. Though perhaps, over time, this will approach a normal distribution as per the [central limit theorem](https://en.wikipedia.org/wiki/Central_limit_theorem).
